Le temps passé sur Instagram varie-t-il selon l’âge ?

```python
temps_instagram_suivant_age= []

data["age"].mean()
age_min = data["age"].min()
age_max = data["age"].max()

print("Âge minimum :", age_min)
print("Âge maximum :", age_max)
```

```python
data["age_group"]= pd.cut(data["age"], bins= [0,18,30,45,66], labels= ["Jeune", "Jeune adulte","Adulte", "Senior"], right=False)
```

```python
data["age_group"]= pd.qcut(data["age"], q=4 , labels= ["Jeune", "Jeune adulte","Adulte","Senior"])

print(data["age_group"].value_counts())
```

```python
temps_instagram_suivant_age= data.groupby("age_group")["daily_active_minutes_instagram"].mean()
print(temps_instagram_suivant_age)
```

```python
temps_instagram_suivant_age.plot(kind="bar", color="blue", figsize=(2,2))
plt.ylabel("Minutes par jour")
plt.title("Temps moyen passé sur Instagram selon l'âge")
plt.show()
```

```python
data["engagement_score"] = (
    data["likes_given_per_day"] +
    data["comments_written_per_day"] +
    data["posts_created_per_week"] +
    data["daily_active_minutes_instagram"] +
    data["dms_sent_per_week"] +
    data["sessions_per_day"] +
    data["time_on_feed_per_day"]
)

engagement_sex = data.groupby("gender")["engagement_score"].mean()

print(engagement_sex)
print("Groupe le plus engagé :", engagement_sex.idxmax())
```

Vérifions avec un t-test

```python
from scipy import stats
```

```python
with_children = data[data["has_children"] == "Yes"]["daily_active_minutes_instagram"]
without_children = data[data["has_children"] == "No"]["daily_active_minutes_instagram"]
```

```python
t_stat, p_value = stats.ttest_ind(with_children, without_children, equal_var=False)

print("Moyenne avec enfants :", with_children.mean())
print("Moyenne sans enfants :", without_children.mean())
print("t_stat :", t_stat)
print("p_value :", p_value)
```
